This notebook can be run in Colab to measure time to run a batch size of 1 on yolov5's multiple model sizes. It was used to generate the results for part 1 of this homework assignment.

```
import torch
from time import time

# Model
model_s = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Images
dir = 'https://github.com/ultralytics/yolov5/raw/master/data/images/'
imgs = [dir + 'bus.jpg']  # pick random image, bus

# Inference, measure time
t = time()
results = model_s(imgs)
time_s = time() - t

# Repeat for other model sizes
model_m = torch.hub.load('ultralytics/yolov5', 'yolov5m', pretrained=True)
t = time()
results = model_m(imgs)
time_m = time() - t

model_l = torch.hub.load('ultralytics/yolov5', 'yolov5l', pretrained=True)
t = time()
results = model_l(imgs)
time_l = time() - t

model_x = torch.hub.load('ultralytics/yolov5', 'yolov5x', pretrained=True)
t = time()
results = model_x(imgs)
time_x = time() - t

# Output results
print('Processed bus image on multiple yolov5 models. Time:')
print(f'Small:  {time_s}')
print(f'Medium: {time_m}')
print(f'Large:  {time_l}')
print(f'XL:     {time_x}')
```

```
from PIL import Image

# Open intersection
img = Image.open('intersection.jpg')

# Run and save results for each model size
s = model_s(img)
s.save('results_s/')
s.print()
m = model_m(img)
m.save('results_m/')
m.print()
l = model_l(img)
l.save('results_l/')
l.print()
x = model_x(img)
x.save('results_x/')
x.print()
```
